```python
from textblob import TextBlob
# Function to calculate sentiment polarity using TextBlob
def get_sentiment(text):
    analysis = TextBlob(text)
    # Return polarity score (ranges from -1 to 1)
    return analysis.sentiment.polarity
```

```python
# Apply sentiment analysis to each text in the dataset
df['Sentiment'] = df['text'].apply(get_sentiment)
```

```python
df['Sentiment']
```

```python
# Define a function to classify polarity scores into positive, negative, or neutral
def classify_sentiment(score):
    if score > 0:
        return 'Positive'
    elif score < 0:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
# Apply classification to sentiment scores
df['Sentiment_Class'] = df['Sentiment'].apply(classify_sentiment)
```
